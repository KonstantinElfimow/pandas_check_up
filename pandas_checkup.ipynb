Группировка категориальных переменных

```python
df.groupby('education_level', dropna=False)['target'].agg(['count', 'mean'])
```

```python
def education_group(x) -> str:
  if x in ['High School', 'Primary School']:
    return 'School'
  if x in ['Masters', 'Phd']:
    return 'Masters_and_Phd'
  return x  
```

```python
df['new_educational_level'] = df['education_level'].apply(education_group)
```

```python
df
```

```python
df.groupby('new_educational_level', dropna=False)['target'].agg(['count', 'mean'])
```

```python
d = {'High School': 'School', 'Primary School': 'School', 'Masters':'Masters_and_Phd', 'Phd': 'Masters_and_Phd', 'Graduate': 'Graduate'}
```

```python
df['education_level'].map(d)
```

```python
def new_training_hours(row):
  if row['education_level'] == 'Phd':
    return row['training_hours'] + 1000
  return row['training_hours']
```

```python
df['new_training_hours'] = df.apply(new_training_hours, axis=1)
```

```python
df
```
